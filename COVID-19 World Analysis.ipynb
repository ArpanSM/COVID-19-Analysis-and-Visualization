# World Coronavirus Analysis

```python
df=pd.read_csv('covid_19_data.csv',parse_dates=['Last Update'])
```

```python
df.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace=True)
```

```python
df
```

Data about UK

```python
df.query('Country=="UK"')
```

```python
T_cases = df['Confirmed'].sum()
print("Total Confirmed: " +str(T_cases))


T_death = df['Deaths'].sum()
print("Total Death: " +str(T_death))

T_recov = df['Recovered'].sum()
print("Total Recovered: " +str(T_recov))

FatalityRate = T_death/T_cases *100
print("Fatality Rate: " +str(FatalityRate)+"%")

RecovRate = T_recov/T_cases *100
print("Recovery Rate Rate: " +str(RecovRate)+"%")
```

```python
#plotly graph objects
fig=go.Figure()
fig.add_trace(go.Bar(x=df['Country'],y=df['Deaths']))
fig.update_layout(title='Total Cases in Country',xaxis=dict(title='Country'),yaxis=dict(title='Deaths'))
```

```python
#plotly graph objects
fig=go.Figure()
fig.add_trace(go.Bar(x=df['Country'],y=df['Recovered']))
fig.update_layout(title='Total Cases in Country',xaxis=dict(title='Country'),yaxis=dict(title='Recovered'))
```

```python
df.iplot(kind='bar',x='Country',y='Confirmed')
```

```python
df.groupby('Date').sum()
```

```python
confirmed=df.groupby('Date').sum()['Confirmed'].reset_index()
death=df.groupby('Date').sum()['Deaths'].reset_index()
rec=df.groupby('Date').sum()['Recovered'].reset_index()
```

Plotting Confirmed cases, deaths and recovery at the same graph

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=2)))

fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=2)))
fig.add_trace(go.Scatter(x=rec['Date'],y=rec['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=2)))
```

# Time Series Analysis of COVID-19 around the world

```python
df_confirmed=pd.read_csv(r'time_series_covid_19_confirmed.csv')
```

```python
df_confirmed.rename(columns={'Country/Region':'Country'},inplace=True)
df_latlong=pd.merge(df,df_confirmed,on=['Country','Province/State'])
df_latlong
```

```python
fig=px.density_mapbox(df_latlong,lat="Lat",lon="Long",hover_name="Province/State",hover_data=["Confirmed","Deaths","Recovered"],animation_frame="Date",color_continuous_scale="Portland",radius=7,zoom=0,height=700)
fig.update_layout(title='Worldwide Corona Virus Cases')
fig.update_layout(mapbox_style="open-street-map",mapbox_center_lon=0)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
```
